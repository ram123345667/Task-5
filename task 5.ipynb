```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r'C:\Users\RAMPRASAD\Downloads\train.csv')  # Change the path as needed
# Overview of the dataset
print(df.info())
print(df.describe())

# Categorical value counts
print(df['Survived'].value_counts())
print(df['Pclass'].value_counts())
print(df['Sex'].value_counts())
print(df['Embarked'].value_counts())
```

```python
#Visualizations & Distributions
sns.set(style="whitegrid")

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Exploratory Data Analysis: Key Distributions', fontsize=16)

# Age distribution
sns.histplot(df['Age'].dropna(), kde=True, bins=30, ax=axs[0, 0], color='skyblue')
axs[0, 0].set_title('Age Distribution')

# Fare distribution
sns.histplot(df['Fare'], kde=True, bins=30, ax=axs[0, 1], color='lightgreen')
axs[0, 1].set_title('Fare Distribution')

# Survival counts
sns.countplot(x='Survived', data=df, ax=axs[1, 0], palette='pastel')
axs[1, 0].set_title('Survival Counts')
axs[1, 0].set_xticklabels(['Not Survived', 'Survived'])

# Passenger class distribution
sns.countplot(x='Pclass', data=df, ax=axs[1, 1], palette='muted')
axs[1, 1].set_title('Passenger Class Counts')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
#Boxplots & Correlation Heatmap
fig, axs = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('Feature Relationships with Survival', fontsize=16)

# Age vs Survival
sns.boxplot(x='Survived', y='Age', data=df, ax=axs[0], palette='Set3')
axs[0].set_title('Age vs Survival')
axs[0].set_xticklabels(['Not Survived', 'Survived'])

# Fare vs Survival
sns.boxplot(x='Survived', y='Fare', data=df, ax=axs[1], palette='Set2')
axs[1].set_title('Fare vs Survival')
axs[1].set_xticklabels(['Not Survived', 'Survived'])

# Correlation heatmap
corr = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', ax=axs[2])
axs[2].set_title('Correlation Heatmap')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python

# Pairplot
# Sample to reduce rendering time if needed
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived', palette='husl')
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```
